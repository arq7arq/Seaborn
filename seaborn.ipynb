seaborn

```python
import pandas as pd
import seaborn as sns
```

```python
housing = pd.read_csv('./housing.csv')
housing.head()
```

## Basic

```python
g = sns.relplot(kind='scatter', data=housing, x='longitude', y='latitude')
g.set(title='Geographical location of housing in california')
```

```python
g = sns.relplot(kind='scatter', data=housing, x='longitude', y='latitude', hue='median_house_value')
g.set(title='Geographical location of housing in california')
```

```python
g = sns.relplot(kind='scatter', data=housing, x='longitude', y='latitude', style='ocean_proximity')
g.set(title='Geographical location of housing in california')
```

```python
g = sns.relplot(kind='scatter', data=housing, x='longitude', y='latitude', style='ocean_proximity', hue='median_house_value')
g.set(title='Geographical location of housing in california')
```
